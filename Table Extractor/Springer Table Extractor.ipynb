```python
#Extracting Title from Springer Sample Article
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET

filepath = "C:/Users/walid/OneDrive/Documents/XML file from Springer sample article.xml"

def header(filepath): #define header as filepath   
    lst = [] #create empty array
    tree = ET.parse(filepath)#read in the XML
    for tags in tree.iter('table-wrap'): #for each tag in interation through 'table-wrap' tag 
        for caption in tags.find('caption'): # for each caption tag
            for p in caption.iter('bold'): # for each p tag nested in caption
                for title in p.iter('bold'): #for each title in 'p' tag
                    name = title.text #Save title text as name
                    lst.append(name) # append name to array named lst
    return lst #end execution and returns everything in 1st array
output = header(filepath) #why is this here
print(output)
```

```python
# Extracting Column titles
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET
filepath = "C:/Users/walid/OneDrive/Documents/XML file from Springer sample article.xml"
tree = ET.parse(filepath)
root = ET.tostring(tree.getroot())
lst = tree.iter('thead') # create list of 'thead' tags
total_list = ['material_names']
headers = [] #empty array

for tags in lst: # for each 'thead' tag inside lst
    rows = tags.findall('tr') # find all 'tr' tags inside specific 'thead' tag currently referenced by variable name tags
    entry_list = rows[0] #starts off with row 0 which is the header 
    column_names = entry_list.findall('th') #finds every 'th' tag in first row
    
    for column_name in column_names:
        values = "" # creates empty string 
        for x in column_name.itertext(): # iterates through every piece of text in the tags that are children of column_name
            values = values + x # For every text in iterator append to 'values' string
        headers.append(values) # once header name is reconstructed for specific column name, add it to list of header names
for x in headers: #go through list of headers and print them out
    print(x)
    
```
